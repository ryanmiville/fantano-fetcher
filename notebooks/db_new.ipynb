```python
file_name = "descriptions_after_2023-06-24.jsonl"
date = "2023-06-24"
```

```python
import pandas as pd

# read in descriptions.jsonl
df = pd.read_json(file_name, lines=True)
df
```

```python
import os
# returns True if video_id is a filename in dataset/yellow folder
def is_yellow(video_id):
    return os.path.isfile(f"thumbnails/yellow/{video_id}.jpg")
```

```python
"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Derive column 'derivedCol' from column: 'title'
    # Transform "title" as per the following examples:
    #   HMLTD - The Worm ALBUM REVIEW ==> HMLTD
    #   Depeche Mode - Memento Mori ALBUM REVIEW ==> Depeche Mode
    df.insert(2, "artist", df["title"].str.split(" -").str[0].str.strip())
    # Derive column 'derivedCol' from column: 'title'
    # Transform "title" as per the following examples:
    #   HMLTD - The Worm ALBUM REVIEW ==> The Worm
    #   Depeche Mode - Memento Mori ALBUM REVIEW ==> Memento Mori
    df.insert(3, "album", df["title"].str.split(" - ").str[1].str.strip().str.removesuffix(" ALBUM REVIEW"))

    df.loc[:, "description"] = df["description"].str.replace("\\n", "\n")
    # na = df[df["album"].isnull()]
    # na = na[na["title"].str.contains("- ")]
    # na.loc[:,"artist"] = na["title"].str.split("- ").str[0]
    # na.loc[:,"album"] = na["title"].str.split("- ").str[1].str.removesuffix(" ALBUM REVIEW")
    # df.update(na)
    
    df["yellow_flannel"] = df["videoId"].apply(is_yellow)
    # rename all columns to be snake_case
    df = df.rename(columns={'videoId': 'video_id', 'thumbnailUrl': 'thumbnail_url', 'publishDate': 'publish_date', 'watchUrl': 'watch_url'})
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
df_clean[df_clean["album"].isnull()]
```

```python
clean_file_name = f"descriptions_clean_after_{date}.jsonl"
```

```python
#write to jsonl file
df_clean.to_json(clean_file_name, orient='records', lines=True)
```

```python
df = pd.read_json(clean_file_name, lines=True)
df['rating'] = df['description'].str.extract(r'\n(\d+)/10')
df.head()
```

```python
broken_ratings = df[df["rating"].isnull()]
# collect description of broken ratings as list[str]
descs = broken_ratings["description"].tolist()
descs

```
